# Restricted Boltzmann Machine

Las Máquinas Restringidas de Boltzmann (RBM) son un tipo de red generativa que ha pesar de su simplicidad son de gran utilidad como estrategia de reducción de dimensionalidad. 

Este tipo de red neuronal consta de 2 capas, la primera capa (capa de entrada) se compone de tantas neuronas como variables de entrada. **Esta capa es también llamada capa visible**. Asimismo, las RBMs constan de una segunda capa, **también conocida como capa oculta**, la cual es en además la capa de salida de la neurona. Por lo tanto, tenemos una matriz de pesos, y bias para las capas visibles y oculta. Entonces, la función de costo consiste en **minimizar la diferencia entre los datos originales y la reconstrucción de estos**.

### Creando el modelo

Ya que RBMs son grafos no dirigidos, no ajustan su peso a traves del descenso del gradiente o backpropagation, sino que hacen uso de un proceso llamado divergencia contrastiva. 

```python
first = X_train[0:1,:]
first.shape
```

```python
import numpy as np

class RBM:
    def __init__(self, n_input_neurons, n_output_neurons):
        self.W = np.zeros((n_input_neurons, n_output_neurons))
        
        self.bv = np.random.rand(n_input_neurons) #visible layer bias
        self.bv = self.bv.reshape((1,-1))
        
        self.bh = np.random.rand(n_output_neurons) #hidden layer bias
        self.bh = self.bh.reshape((1,-1))
    
    def sample(self, probability_distribution):
        random_dist = np.random.uniform(0,1, probability_distribution.shape)
        
        example = probability_distribution - random_dist
        example[example<0.0] = 0.0
        example[example>0.0] = 1.0
        return example
        
    def encode(self, X):
        probability_distribution = functions.sigm[0](X @ self.W + self.bh) #data distribution probability
        return self.sample(probability_distribution)
        
    def decode(self, X):
        probability_distribution = functions.sigm[0](X @ self.W.T + self.bv) #data distribution probability
        return self.sample(probability_distribution)
    
    def getReconstructedOutput(self, X):
        encode = self.encode(X)
        decode = self.decode(encode)
        return decode
        
    def train(self, X, loss_function, lr=.01, epochs=1000, verbose = False):
        epoch = 0
        history = []
        while epoch < epochs:
            epoch+=1
            
            h0 = self.encode(X)
            v1 = self.decode(h0)
            h1 = self.encode(v1)

            delta_W = X.T @ h0 - v1.T @ h1
            self.W = self.W + lr * delta_W
            
            self.bv = self.bv + lr *  np.mean(functions.l2_cost[1](v1 , X), axis=0, keepdims=True)
            self.bh = self.bh + lr * np.mean(functions.l2_cost[1](h0 , h1), axis=0, keepdims=True)
                
            loss = loss_function[0](v1, X) #loss
            history.append(loss)
            if verbose:
                print(f'Epoch {epoch} ==> Loss: {loss}')
        return history
```

```python
neural_net = RBM(64, 63)
```

```python
history = neural_net.train(X_train, functions.l2_cost, verbose = True)
plt.plot(history)
```

```python
reconstructed = neural_net.getReconstructedOutput(test)

plt.imshow(reconstructed.reshape((1,8,8))[0], cmap=plt.cm.gray_r, interpolation="nearest")
```
