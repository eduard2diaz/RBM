```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from sklearn.decomposition import PCA
```

```python
img = mpimg.imread('./original.png')
print(img.shape)
plt.imshow(img)
```

```python
img_first_channel = img[:,:,0]
plt.imshow(img_first_channel, cmap='gray')
```

```python
import numpy as np

def reduction(img, n_components = -1):
    if n_components == -1:
        pca = PCA()
    else:
        pca = PCA(n_components)
        
    reduced = pca.fit_transform(img)
    reduced_recovered = pca.inverse_transform(reduced)
    
    percentage = np.cumsum(pca.explained_variance_ratio_ * 100)[-1]
    
    fig, ax = plt.subplots()
    ax.imshow(reduced_recovered, cmap='gray')
    plt.show()
    return percentage
```

```python
percentage = 0
n_components = 1
variance_percentage_limit = 99
while percentage < variance_percentage_limit:
    percentage = reduction(img_first_channel, n_components)
    print(f'N_components: {n_components} Variance porcentaje {percentage}%')
    n_components+= 1
```
